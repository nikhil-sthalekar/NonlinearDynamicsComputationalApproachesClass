```python
from pynamical import logistic_map, simulate, bifurcation_plot
%matplotlib inline
from numba import jit
```

```python
@jit(nopython=True)
def log_map(pop, rate):
    return rate * pop * (1 - pop)
```

```python
pops = simulate(model=log_map, num_gens=1000, rate_min=0, rate_max=4, num_rates=1000, num_discard=500)
```

```python
bifurcation_plot(pops, xmin=3.84, xmax=3.8571, ymin=0.4, ymax=0.6)
```
